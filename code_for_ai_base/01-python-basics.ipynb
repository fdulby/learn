```python
# def with default input value

import math

def calculate_size(r, h, display=True):
    s = math.pi * r * r * h
    if display:
        print('r = {:.2f}, h = {:.2f} -> s = {:.2f}'.format(r, h, s))
    return s

r1 = 1.0
h1 = 2.0
s1 = calculate_size(r1, h1)

r2 = 3.0
h2 = 4.0
s2 = calculate_size(r2, h2, display=False)
print(s2)
```

```python
## class

# Cylinder

import math

class Cylinder:
    def __init__(self, r=0, h=0):
        self.r = r
        self.h = h
        print('cylinder object created')

    def calculate_size(self, display=True):
        s = math.pi * self.r * self.r * self.h
        if display:
            print('r = {:.2f}, h = {:.2f} -> s = {:.2f}'.format(self.r, self.h, s))
        return s
    
    def is_larger(self, other):
        if not isinstance(other, Cylinder):
            print('other must be a Cylinder')
            return None
        if self.calculate_size(False) > other.calculate_size(False):
            return True
        else:
            return False

c1 = Cylinder(r=1.0, h=2.0)
print(c1.r, c1.h)
s1 = c1.calculate_size()

c2 = Cylinder(3.0, 4.0)
print(c2.r, c2.h)
s2 = c2.calculate_size()

print(c1.is_larger(c2))
print(c1.is_larger(1))
```

```python
# Cuboid and Cylinder

import math

class Cuboid:
    def __init__(self, x=0, y=0, z=0):
        self.x = x
        self.y = y
        self.z = z
        print('cuboid object created')
        
    def calculate_size(self, display=True):
        s = self.x * self.y * self.z
        if display:
            print('x = {:.2f}, y = {:.2f}, z = {:.2f} -> s = {:.2f}'.format(self.x, self.y, self.z, s))
        return s
        
    def is_larger(self, other):
        if self.calculate_size(False) > other.calculate_size(False):
            return True
        else:
            return False

class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
        print('cylinder object created')

    def calculate_size(self, display=True):
        s = math.pi * self.r * self.r * self.h
        if display:
            print('r = {:.2f}, h = {:.2f} -> s = {:.2f}'.format(self.r, self.h, s))
        return s
    
    def is_larger(self, other):
        if self.calculate_size(False) > other.calculate_size(False):
            return True
        else:
            return False

o1 = Cuboid(1.0, 2.0, 3.0)
s1 = o1.calculate_size()

o2 = Cylinder(1.0, 2.0)
s1 = o2.calculate_size()

print(o1.is_larger(o2))
```

```python
# Object, Cuboid and Cylinder

import math

class Object:
    def __init__(self):
        pass
        
    def calculate_size(self, display=True):
        return None
        
    def is_larger(self, other):
        if self.calculate_size(False) > other.calculate_size(False):
            return True
        else:
            return False

    @staticmethod
    def compare_size(obj1, obj2):
        if obj1.calculate_size(False) > obj2.calculate_size(False):
            return True
        else:
            return False

class Cuboid(Object):
    def __init__(self, x=0, y=0, z=0):
        self.x = x
        self.y = y
        self.z = z
        print('cuboid object created')
        
    def calculate_size(self, display=True):
        s = self.x * self.y * self.z
        if display:
            print('x = {:.2f}, y = {:.2f}, z = {:.2f} -> s = {:.2f}'.format(self.x, self.y, self.z, s))
        return s

class Cylinder(Object):
    def __init__(self, r, h):
        self.r = r
        self.h = h
        print('cylinder object created')

    def calculate_size(self, display=True):
        s = math.pi * self.r * self.r * self.h
        if display:
            print('r = {:.2f}, h = {:.2f} -> s = {:.2f}'.format(self.r, self.h, s))
        return s

o1 = Cuboid(1.0, 2.0, 3.0)
s1 = o1.calculate_size()

o2 = Cylinder(1.0, 2.0)
s1 = o2.calculate_size()

print(o1.is_larger(o2))

print(Object.compare_size(o1, o2))
```
