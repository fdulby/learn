# PyTorch Tensor

```python
import torch
import numpy as np
```

## Initializing a Tensor

```python
# From Python data

data = 1

x = torch.tensor(data)  # torch.int64

print(x.shape)
print(x.dim())
print(x.dtype)

print('x =', x)
print('x.data =', x.data)
print('x.grad =', x.grad)

print(x.item(), type(x.item()))

print(x.tolist(), type(x.tolist()))
```

```python
# From Python list

data = [1., 2.]

x = torch.tensor(data, requires_grad=True)  # turn on autograd; torch.float32

print(x.shape)
print(x.dim())
print(x.dtype)

print('x =', x)
print('x.data =', x.data)
print('x.grad =', x.grad)

# print(x.item(), type(x.item()))  # Error: a Tensor with 4 elements cannot be converted to Scalar

print(x.tolist(), type(x.tolist()))
```

```python
# From Python list (set dtype)

data = [1, 2]

x = torch.tensor(data, dtype=torch.float)

print(x.shape)
print(x.dim())
print(x.dtype)

print('x =', x)
print('x.data =', x.data)
print('x.grad =', x.grad)
```

```python
# From NumPy array

data = [[1., 2.], [3., 4.]]

data = np.array(data)  # float64

x = torch.tensor(data)

print(x.shape)
print(x.dim())
print(x.dtype)

print('x =', x)
print('x.data =', x.data)
print('x.grad =', x.grad)

data[0, 0] = 0.
print(data)
print(x)
```

```python
# From NumPy array

data = [[1., 2.], [3., 4.]]

data = np.array(data)  # float64

x = torch.from_numpy(data)

print('x =', x)
print('x.data =', x.data)
print('x.grad =', x.grad)

data[0, 0] = 0.
print(data)
print(x)
```

```python
# To NumPy array

x = torch.ones((3, 4))

print('x =', x)
print('x.data =', x.data)
print('x.grad =', x.grad)

n = x.numpy()
print(x)

x[0] = 0
print(x)
print(n)

n[0] = 1
print(x)
print(n)
```

```python
# To NumPy array

x = torch.ones((3, 4), requires_grad=True) # turn on autograd

print('x =', x)
print('x.grad =', x.grad)

# n = x.numpy()  # Error: Can't call numpy() on Tensor that requires grad.

n = x.detach().numpy()
```

## Operations on Tensors

```python
# Standard numpy-like indexing and slicing

tensor = torch.ones((4, 4))
print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")

tensor[:, 1] = 0
print(tensor)
```

```python
# Joining tensors

tensor = torch.ones((4, 4))

xx = torch.cat([x, x], dim=1)

print(xx.shape)

print(xx)
```

```python
# Arithmetic operations

tensor = torch.ones((4, 4))

y1 = x @ x.T
print(y1)

y2 = x.matmul(x.T)
print(y2)
```

```python
x = torch.rand(3, 4)

agg = x.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))

agg = x.mean()
agg_item = agg.item()
print(agg_item, type(agg_item))

agg = x.max()
agg_item = agg.item()
print(agg_item, type(agg_item))
```
