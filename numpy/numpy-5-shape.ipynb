### 数组形状操作

```python
import numpy as np
tang_array = np.arange(10)
tang_array
```

```python
tang_array.shape
```

```python
tang_array.shape = 2,5
#将一个10数字的序列构造为2，5的格式
```

```python
tang_array
```

```python
tang_array.reshape(1,10)
```

大小必须不能改变

```python
tang_array .shape = 3,4
```

```python
tang_array = np.arange(10)
tang_array.shape
```

```python
tang_array = tang_array[np.newaxis,:]
#新增一个轴
tang_array.shape
```

```python
tang_array = np.arange(10)
tang_array.shape
```

```python
tang_array = tang_array[:,np.newaxis]
tang_array.shape
```

```python
tang_array = tang_array[:,np.newaxis,np.newaxis]
tang_array.shape
```

```python
tang_array = tang_array.squeeze()
#将多余的轴做一个压缩
tang_array.shape
```

```python
tang_array.shape = 2,5
```

```python
tang_array
```

```python
tang_array.transpose()
#转置
```

```python
tang_array.T
```

```python
tang_array
```

### 数组的连接

```python
a = np.array([[123,456,678],[3214,456,134]])
a
```

```python
b = np.array([[1235,3124,432],[43,13,134]])
b
```

```python
c = np.concatenate((a,b))
#默认按照第一个维度
c
```

```python
c = np.concatenate((a,b),axis = 0)
c
```

```python
c = np.concatenate((a,b),axis = 1)
c
```

```python
c.shape
```

```python
np.vstack((a,b))
#vstack就是表示在轴等于0，竖着拼接
```

```python
np.hstack((a,b))
#hstack就是横着拼接
```

```python
a
```

```python
a.flatten()
#直接拉平
```
