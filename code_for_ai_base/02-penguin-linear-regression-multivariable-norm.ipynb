# Penguin Multivariable Linear Regression with Data Normalization
带有数据归一化的线性回归

```python
# read the variables: class label and features

def string2float(item):
    return float(item) if item != '' else None

def read_strings(filename, col):
    with open(filename) as f:
        lines = f.readlines()
    values = [line.strip().split(',')[col] for line in lines[1:]]
    return values

def read_floats(filename, col):
    with open(filename) as f:
        lines = f.readlines()
    values = [string2float(line.strip().split(',')[col]) for line in lines[1:]]
    return values
```

```python
import numpy as np

filename = 'data/palmer-penguins/palmer-penguins-train.txt'

x1 = np.array(read_floats(filename, col=2))
x2 = np.array(read_floats(filename, col=3))
x3 = np.array(read_floats(filename, col=4))

y = np.array(read_floats(filename, col=5))
```

```python
idx = (x1 != None) & (x2 != None) & (x3 != None) & (y != None)

x1 = x1[idx]
x2 = x2[idx]
x3 = x3[idx]

y = y[idx]
```

```python
# Min-Max
#拼接x1，x2，x3
X = np.stack((x1, x2, x3), axis=1)

X_min = np.min(X, axis=0)
X_max = np.max(X, axis=0)

print(X_min)
print(X_max)
print(np.finfo(float).eps)

X = (X - X_min)/(X_max - X_min + np.finfo(float).eps)
print(X[:10])
```

```python
x0 = np.ones((x1.shape[0], 1))
print(x0.shape)
#加上偏置
print(X.shape)

X = np.concatenate((x0, X), axis=1)
print(X.shape)
print(X[:10])
```

```python
result = np.matmul(X.transpose(), X)
result = np.linalg.inv(result)
result = np.matmul(result, X.transpose())
result = np.matmul(result, y)
w = result

print(w)
```

```python
# training set evaluation

y_pred = np.matmul(X, w)

diff = y - y_pred
mse = np.mean(diff ** 2)

print(mse)
print(np.mean(y), np.sqrt(mse), np.sqrt(mse)/np.mean(y))
```
