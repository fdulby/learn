##这里必须要讲解一下if与while的区别：

if循环是进行一次（条件判断），如果条件满足，就会执行，如果不满足就不执行


但是对于while来说，它将会反复判断，直到不满足的时候才会停下来

```python
# while

animals = ('cat', 'dog', 'fish')

#注意下面的idx +=1的操作，实际上如果没有这一步的话，while将会是死循环
idx = 0
while idx < len(animals):
    print(animals[idx])
    idx += 1
```

注意break是直接在当前的那一行跳出当前所在的这一层循环。如果有多层循环嵌套，它只会退出最里面的那一层，外层的循环会继续执行

```python
# while

animals = ('cat', 'dog', 'fish')

idx = 0
while idx < len(animals):
    if idx < 2:
        print(animals[idx])
    else:
        break
    idx += 1

#下面是其他例子
password = "123456"
attempts = 3

while attempts > 0:                    # while：给 3 次机会（循环）
    guess = input("请输入密码：")

    if guess == password:              # if：判断对错（分支）
        print("登录成功")
        break                          # 跳出 while 循环
    else:
        attempts -= 1
        if attempts > 0:               # if 嵌套：提示还剩几次
            print(f"密码错误，还剩 {attempts} 次机会")
        else:
            print("账户已锁定")
```

下面的def中为什么有一个display=true？
#大多数情况下，用户想"算完顺便看一眼结果" → 默认打印（要么没有display=true，填上空白，要么就是display=true）
#特殊情况下（如批量计算、写入文件），用户不想看到满屏输出 → 可以关掉（这时候可以用display=false）

```python
# def with default input value

import math

def calculate_size(r, h, display=True):
    s = math.pi * r * r * h
    if display:
        print('r = {:.2f}, h = {:.2f} -> s = {:.2f}'.format(r, h, s))
    return s

r1 = 1.0
h1 = 2.0
s1 = calculate_size(r1, h1)

r2 = 3.0
h2 = 4.0
s2 = calculate_size(r2, h2, display=False)
print(s2)
```

### class
下面有class的基本流程，首先是init初始化，然后是各类函数

```python
# Cylinder
#self 是指向对象自身的引用

class Cylinder:
    def __init__(self, r=0, h=0):
        #这里设置的是特征，也是在初始化
        self.r = r
        self.h = h
        print('cylinder object created')
#这里的def中传入两个参数，一个是self这个类，还有一个是是否display
    def calculate_size(self, display=True):
        s = math.pi * self.r * self.r * self.h
        if display:
            print('r = {:.2f}, h = {:.2f} -> s = {:.2f}'.format(self.r, self.h, s))
        return s
    
    def is_larger(self, other):
        if not isinstance(other, Cylinder):
            #isinstance(obj, Class)：检查 obj 是否是 Class 的实例
            print('other must be a Cylinder')
            return None
        #注意这里的Flase出现是因为不想打印出来
        if self.calculate_size(False) > other.calculate_size(False):
            return True
        else:
            return False

c1 = Cylinder(r=1.0, h=2.0)
print(c1.r, c1.h)
s1 = c1.calculate_size()

c2 = Cylinder(3.0, 4.0)
print(c2.r, c2.h)
s2 = c2.calculate_size()

print(c1.is_larger(c2))
print(c1.is_larger(1))
#这里的1就不是一个实例
```

```python
# Cuboid and Cylinder
#这里的用法更加灵活，比较了一个立方体与柱体
import math

class Cuboid:
    def __init__(self, x=0, y=0, z=0):
        self.x = x
        self.y = y
        self.z = z
        print('cuboid object created')
        
    def calculate_size(self, display=True):
        s = self.x * self.y * self.z
        if display:
            print('x = {:.2f}, y = {:.2f}, z = {:.2f} -> s = {:.2f}'.format(self.x, self.y, self.z, s))
        return s
        
    def is_larger(self, other):
        if self.calculate_size(False) > other.calculate_size(False):
            return True
        else:
            return False

class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
        print('cylinder object created')

    def calculate_size(self, display=True):
        s = math.pi * self.r * self.r * self.h
        if display:
            print('r = {:.2f}, h = {:.2f} -> s = {:.2f}'.format(self.r, self.h, s))
        return s
    
    def is_larger(self, other):
        if self.calculate_size(False) > other.calculate_size(False):
            return True
        else:
            return False

o1 = Cuboid(1.0, 2.0, 3.0)
s1 = o1.calculate_size()

o2 = Cylinder(1.0, 2.0)
s1 = o2.calculate_size()

print(o1.is_larger(o2))
```

```python
# Object, Cuboid and Cylinder
import math

class Object:
    def __init__(self):
        pass
    #pass就是用来占位置的
    def calculate_size(self, display=True):
        return None
        
    def is_larger(self, other):
        if not isinstance(other, Object):
            print('other must be an Object')
            return None
        if self.calculate_size(False) > other.calculate_size(False):
            return True
        else:
            return False

#这里就是一个经典的父类继承
#下面的Cuboid继承了Object的所有，所以在最后可以有比较函数is larger
#那问题是两者的calculate不一样怎么办，python的查找规则是先找自己，再找爹

class Cuboid(Object):
    def __init__(self, x=0, y=0, z=0):
        self.x = x
        self.y = y
        self.z = z
        print('cuboid object created')
        
    def calculate_size(self, display=True):
        s = self.x * self.y * self.z
        if display:
            print('x = {:.2f}, y = {:.2f}, z = {:.2f} -> s = {:.2f}'.format(self.x, self.y, self.z, s))
        return s

class Cylinder(Object):
    def __init__(self, r, h):
        self.r = r
        self.h = h
        print('cylinder object created')

    def calculate_size(self, display=True):
        s = math.pi * self.r * self.r * self.h
        if display:
            print('r = {:.2f}, h = {:.2f} -> s = {:.2f}'.format(self.r, self.h, s))
        return s

o1 = Cuboid(1.0, 2.0, 3.0)
s1 = o1.calculate_size()

o2 = Cylinder(1.0, 2.0)
s1 = o2.calculate_size()

print(o1.is_larger(o2))
```

```python
# 1. 打开文件（返回文件对象）
f = open('文件名', '模式')

#这里的模式大有说法
#'r'代表最简单的只读
#‘w'就是写，文件不在的时候就创建，有的时候就会删掉重写

# 2. 读写操作
content = f.read()  # 或 f.write(), f.readline() 等

# 3. 关闭文件（必须！释放系统资源）
f.close()
```

下面是读取的部分

```python
# 方法 1：一次性读全部（小文件适用）
f = open('data.txt', 'r')
content = f.read()          # 返回字符串
f.close()

# 方法 2：逐行读取（大文件推荐，省内存）
f = open('data.txt', 'r')
for line in f:              # 文件对象是可迭代器
    print(line.strip())     # strip() 去掉行尾换行符
f.close()

# 方法 3：读取一行
f = open('data.txt', 'r')
first_line = f.readline()   # 读取一行（包含换行符 \n）
second_line = f.readline()  # 继续读下一行
f.close()

# 方法 4：读取所有行到列表
f = open('data.txt', 'r')
lines = f.readlines()       # 返回 ['第一行\n', '第二行\n', ...]
f.close()
```

然后是写入

```python
# 覆盖写入（w 模式会清空原文件）
f = open('output.txt', 'w')
f.write('Hello World\n')    # 注意手动加换行符
f.write('第二行内容\n')
f.close()

# 追加写入（a 模式，不会清空）
f = open('log.txt', 'a')
f.write('新增日志\n')
f.close()

# 写入多行
lines = ['Line 1\n', 'Line 2\n', 'Line 3\n']
f = open('output.txt', 'w')
f.writelines(lines)         # 不会自动加换行符，需自己处理
f.close()
```

```python
file_name = 'README.md'

f = open(file_name, 'r')

line = f.readline()
print(line)

line = f.readline().rstrip()
print(line)

f.close()
```

```python
file_name = 'README.md'
#然后是采用with语法
#离开with自动会关掉

with open(file_name) as f:
    lines = f.readlines()
    print(lines)

print(lines)

print(len(lines))

#rstrip() 是 Python 字符串的右侧清理方法，用于移除字符串右边（末尾）的指定字符，默认移除空白字符（包括空格、换行符 \n、制表符 \t 等）。

for idx in range(len(lines)):
    print('line', idx, ':', lines[idx].rstrip())
```

```python
file_name = 'README.md'

with open(file_name) as f:
    lines = f.readlines()

for line in lines:
    print(line.rstrip())
```

```python
file_name = 'README.md'

with open(file_name) as f:
    lines = f.readlines()

for idx, line in enumerate(lines):
    print(idx, line.rstrip())
```
