# PyTorch Backpropagation and Gradient Calculation

```python
import torch
```

```python
# data generation

x = torch.tensor([1., 2., 3.], requires_grad=False)
y = torch.tensor(36., requires_grad=False)
w = torch.tensor([4., 5., 6.], requires_grad=True)
```

```python
print('x      =', x)
print('x.data =', x.data)
print('x.grad =', x.grad)
print('x.if_leaf =', x.is_leaf)

print()

print('y      =', y)
print('y.data =', y.data)
print('y.grad =', y.grad)
print('y.if_leaf =', y.is_leaf)

print()

print('w      =', w)
print('w.data =', w.data)
print('w.grad =', w.grad)
print('w.if_leaf =', w.is_leaf)
```

```python
# dot product calculation

p = torch.dot(w, x)

print('p      =', p)
print('p.data =', p.data)
print('p.grad =', p.grad)
print('p.if_leaf =', p.is_leaf)
print('p.grad_fn =', p.grad_fn)
```

```python
print('p:', p.grad_fn)
print(p.grad_fn.next_functions)
print(p.grad_fn.next_functions[0][0].next_functions)

print(p.grad_fn._saved_self)

print('')

print('w:', w.grad_fn)
print('x:', x.grad_fn)
```

```python
# loss calculation

l = (p - y) ** 2

print('l      =', l)
print('l.data =', l.data)
print('l.grad =', l.grad)
print('l.if_leaf =', l.is_leaf)
print('l.grad_fn =', l.grad_fn)
```

```python
print('l.grad_fn:', l.grad_fn)  # <PowBackward0 object at 0x7f4370ea7b50>
print('l.grad_fn._saved_self:', l.grad_fn._saved_self)  # tensor(-4., grad_fn=<SubBackward0>)
print('l.grad_fn._saved_self.grad_fn:', l.grad_fn._saved_self.grad_fn)  # <SubBackward0 object at 0x7f4370c266e0>
print('l.grad_fn.next_functions:', l.grad_fn.next_functions)  # ((<SubBackward0 object at 0x7f4370c266e0>, 0),)
print(l.grad_fn.next_functions[0][0].next_functions)  # ((<DotBackward0 object at 0x7f4370e1bc10>, 0), (None, 0))

print()

print('p.grad_fn:', p.grad_fn)  #  <DotBackward0 object at 0x7f973a4833d0>
print('p.grad_fn.next_functions:', p.grad_fn.next_functions)  # ((<AccumulateGrad object at 0x7f973a482bc0>, 0), (None, 0))

print()

print('w:', w.grad_fn)
print('x:', x.grad_fn)
```

```python
# back prop

l.backward()

print('After backward:')

print('p.grad =', w.grad)
print('x.grad =', x.grad)
print('w.grad =', w.grad)
```

```python
p = torch.dot(w, x)
l = (p - y) ** 2
l.retain_grad()
l.backward()

print('After backward:')

print('p.grad =', w.grad)
print('x.grad =', x.grad)
print('w.grad =', w.grad)
```
