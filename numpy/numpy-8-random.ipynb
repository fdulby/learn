### Numpy的随机模块

```python
import numpy as np
# rand: 均匀分布 [0, 1)，参数是各维度大小
# randn: 标准正态分布（均值0，标准差1）
# randint: 随机整数 [low, high)
np.random.rand(3,2)
```

```python
#返回的是随机的整数，左闭右开
np.random.randint(10,size = (5,4))
```

```python
np.random.rand()
```

```python
np.random.random_sample()
```

```python
np.random.randint(0,10,3)
```

```python
#对于随机的高斯分布来说，mu为均值，sigma为标准差
mu, sigma = 0,0.1
np.random.normal(mu,sigma,10)
```

```python
#可以指定精度
np.set_printoptions(precision = 2)
```

```python
mu, sigma = 0,0.1
np.random.normal(mu,sigma,10)
```

```python
# 正态分布（自定义均值和标准差）
np.random.normal(loc=0, scale=1, size=(3, 3))     # 均值0，标准差1
np.random.normal(100, 15, 1000)                   # 1000个N(100,15²)

# 均匀分布（自定义范围）
np.random.uniform(low=0, high=10, size=5)         # [0,10)均匀分布
```

### 随机的种子

```python
#生成随机数的时候，在没有种子的情况下，每个数都是以不同的方式随机的,那么每一次生成的结果都不一样
#计算机生成的随机数实际上是伪随机的（Pseudo-random）。它们由确定性的数学算法生成，种子就是这个算法的初始值。
#相同种子 → 完全相同的随机数序列
np.random.seed(100)
np.random.rand(3)
```

```python
mu, sigma = 0,0.1
np.random.normal(mu,sigma,10)
```

```python
# choice: 从一维数组中随机抽取
np.random.choice([10, 20, 30], size=3)      # 抽取3个元素
np.random.choice(100, size=5, replace=False) # 0-99不重复抽取5个
np.random.choice(['a', 'b', 'c'], size=10, p=[0.5, 0.3, 0.2])  # 带权重抽取
```

```python
# 创建随机数生成器实例（推荐写法）
rng = np.random.default_rng(seed=42)

# 所有方法通过 rng 调用
rng.random((3, 2))            # 等同 rand，但参数需用元组
rng.standard_normal((3, 2))   # 等同 randn
rng.integers(0, 10, (3, 2))   # 等同 randint
rng.choice([1, 2, 3], size=2)
rng.shuffle(arr)
```
