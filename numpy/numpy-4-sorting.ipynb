```python
import numpy as np
tang_array = np.array([[1.5,1.3,7.5],
                      [5.6,7.8,1.2]])
tang_array
```

```python
np.sort(tang_array)
#sort默认从小到大排序，并且是从最后一个轴排序
```

```python
np.sort(tang_array,axis = 0)
```

```python
tang_array
```

```python
np.argsort(tang_array)
#arg仍然是索引位置
```

```python
tang_array = np.linspace(0,10,10)
#从0开始构造10个数，每两的差一样
tang_array
```

```python
values = np.array([2.5,6.5,9.5])
```

```python
np.searchsorted(tang_array,values)
#为什么是3，6，9，解释如下：首先searchsort表示的是我在原数列中插入上述的2.5，6.5，9.5这些值，要求是必须先排序好
```

```python
tang_array = np.array([[1,0,6],
                       [1,7,0],
                       [2,3,1],
                       [2,4,0]])
tang_array
```

```python
index = np.lexsort([-1*tang_array[:,0],tang_array[:,2]])
#lexsort也是排序
index
```

```python
tang_array = tang_array[index]
tang_array
```
